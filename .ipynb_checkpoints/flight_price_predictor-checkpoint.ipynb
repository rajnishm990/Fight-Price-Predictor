### Feature Engineering and Exploratory Data Analysis

#### Duration 

```python
##Handling the Duration part .... as the data is in format like H , M format we have to convert it to hours , we can do this creating a custom function

def convert_duration(column):
    if len(column.split()) == 2:
        hours = int(column.split()[0] [ : -1])
        minutes = int(column.split()[1] [:  -1])
        return hours * 60 + minutes
    else:
        return int(column[ : -1])*60
        
```

```python
## applying this function to every row  of the duration column
df['Duration']=df['Duration'].apply(convert_duration)
df.head()
```

#### Departure And Arrival Time

```python
##handing Departure time and Arrival time 
df['Dep_Time'] = pd.to_datetime(df.Dep_Time)
df['Arrival_Time'] = pd.to_datetime(df.Arrival_Time)
print(df['Dep_Time'].dtype)
print(df['Arrival_Time'].dtype)
```

Since we have now  converted the Departure time and Arrival time in date time format , we can use methods associated with them 


```python
df['dep_time_in_hours'] = df['Dep_Time'].dt.hour
df['dep_time_in_minutes'] = df['Dep_Time'].dt.minute
df['arrival_time_in_hours'] = df['Arrival_Time'].dt.hour
df['arrival_time_in_minutes'] = df['Arrival_Time'].dt.minute
```

```python
df.head()
```

Since we have converted the Arrival and Departure into hours and minutes we don't actually need the Arrival Time and Departure time Columns. Therefore we can drop them 

```python
df.drop(['Arrival_Time', 'Dep_Time'], axis=1, inplace=True)
df.head()
```

####  Date of Journey

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])

df.head()
```

 If we look closely at the date part we can deduce that we only have the data of the year 2019

```python
df['Date_of_Journey'].dt.year.unique()
```

So we don't actually need the year part , we can extract the day and month and drop the year part 

```python
df['Day'] = df['Date_of_Journey'].dt.day
df['Month'] = df['Date_of_Journey'].dt.month
df.drop('Date_of_Journey', inplace=True, axis = 1)
df.head()
```

#### Total Stops

Total Stops column is a ctergorical column and we can map each of its unique value to some integers

```python

df['Total_Stops'].value_counts()
```

So we have 5 values each corresponding to number of stops , hence we can map 1 stop to 1 , non stop to 0 and so on

```python
df['Total_Stops'] = df['Total_Stops'].map({
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4
})
df.head()
```

Now we concatenate the dummy variables to our original DF and drop the original Airline column

```python
df = pd.concat([df,Airline], axis=1)
df.head()
```

```python
df.drop('Airline' , axis=1, inplace=True)
df.head()
```

#### Source and Destination
As both of them are categorical , we have to encode them too 


```python
df.head()
```

```python
df = pd.get_dummies(data=df , columns =['Source', 'Destination'], drop_first=True)
df.head()
```

#### Routes

The route has a arrow which shows the direction of the airline and all the stops it took , we can change this by making seperate columns for all the routes 

```python
route = df[['Route']]
route['Route_1'] = route['Route'].str.split('→').str[0]
route['Route_2'] = route['Route'].str.split('→').str[1]
route['Route_3'] = route['Route'].str.split('→').str[2]
route['Route_4'] = route['Route'].str.split('→').str[3]
route['Route_5'] = route['Route'].str.split('→').str[4]
route.head()
```

```python
## REplacing all the NA values with None
route.fillna('None', inplace =True)
route.head()
```

Since there are many values that need to be encoded , one hot encoder won't be a good choice as it will create many dummy variables. So we use Label Encoer

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for i in range(1,6):
    col = 'Route_'+str(i)
    route[col] = encoder.fit_transform(route[col])

route.head()
```

```python
route.drop('Route', axis = 1, inplace=True)
```

```python
df = pd.concat([df, route], axis=1)
df.head()
```

```python
df.drop('Route', axis =1 , inplace=True)
df.head()
```

All Of our Preprocessing steps are done now and hence we can now move on to the Model fitting and predcition part 

## Prediction 

Before feeding our data into out model to fit the model we have to perform Scaling of the data to a certain level so that all of the column hold equal weight in our prediction stage irrespective of their quantity (For e.g our model might think that duration takes holds more weight in our model because it is in 3 figures than a hour quantity. But we know its not true as duration is in minutes  )

From the data frame it is clear that is price is the dependent column(i.e it is to be precited ) and hence we can safely split the data into X and y where X contains all the indpendent columns and y contain dependent column(price)

```python
y = df['Price']
X = df.drop('Price', axis=1 )
```

```python
## Scaling the data 
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()
X= sc.fit_transform(X)
print(X)
print(y)
```

Now we can see the data has been scaled to the values between 1 to -1

```python
##Splitting the data into training and test set 
from sklearn.model_selection import train_test_split
X_train , X_test, y_train,y_test = train_test_split(X,y, test_size=0.1 , random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error , r2_score

def metric(y_true, y_pred):
    print(f"Mean Squared Error : ", mean_squared_error(y_true,y_pred) ** 0.05 )
    print(f"R2 Score : ", r2_score(y_true , y_pred))

def accuracy(y_true, y_pred):
    errors = abs(y_true - y_pred)
    mape = 100 * np.mean(errors/y_true)
    accuracy= 100-mape
    return accuracy
```

```python
y_pred = regressor.predict(X_test)
```

```python
metric(y_test, y_pred)
```

```python
accuracy(y_test , y_pred)
```

So Accuracy for our Simple Linear Rgression model is 76 %
